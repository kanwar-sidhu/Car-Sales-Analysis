## 9.Monthly sales of cars 

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year_month'] = df['Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby(['Company', 'year_month'])['Price ($)'].sum().reset_index()
```

```python
monthly_sales.head(50)
```

```python
chevrolet_sales = monthly_sales[monthly_sales['Company'] == 'Chevrolet']
```

```python
chevrolet_sales
```

## Chevrolt cars sales monthly

```python
chevrolet_sales = monthly_sales[monthly_sales['Company'] == 'Chevrolet'].copy()
chevrolet_sales['year_month'] = chevrolet_sales['year_month'].apply(lambda x: x.to_timestamp())

chevrolet_sales = chevrolet_sales.sort_values('year_month')

plt.figure(figsize=(10, 6))
plt.plot(chevrolet_sales['year_month'], chevrolet_sales['Price ($)'], marker='o')

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.title('Chevrolet Car Sales Over Time')
plt.xlabel('Year and Month')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

### Most and Least selling models in all Dataset

```python
fig,ax=plt.subplots(1,2,figsize=(10,4))

top_mod=df.Model.value_counts().head(10)
least_mod=df.Model.value_counts().tail(10)

top_mod.plot(kind='bar',color='g',ax=ax[0])
ax[0].set_title('Most selling models')
ax[0].set_ylabel('Count')

least_mod.plot(kind='bar',color='y',ax=ax[1])
ax[1].set_title('least selling models')
ax[1].set_ylabel('Count')

plt.show()
```

### Most and least company sellings 

```python
fig,ax=plt.subplots(1,2,figsize=(10,3))

top_comp=df.Company.value_counts().head(10)
least_comp=df.Company.value_counts().tail(10)

top_comp.plot(kind='bar',color='b',ax=ax[0])
ax[0].set_title('Most selling companies')
ax[0].set_ylabel('Count')

least_comp.plot(kind='bar',color='c',ax=ax[1])
ax[1].set_title('least selling companies')
ax[1].set_ylabel('Count')

plt.show()
```
